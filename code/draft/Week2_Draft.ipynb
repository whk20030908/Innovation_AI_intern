```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv("../../data/raw/etf_prices.csv")
prices["Date"] = pd.to_datetime(prices["Date"])
prices = prices.set_index("Date")

returns = pd.read_csv("../../data/raw/etf_returns.csv")
returns["Date"] = pd.to_datetime(returns["Date"])
returns = returns.set_index("Date")

print(prices.head())
print(returns.head())
print(prices.shape)
print(returns.shape)
```

```python
print(prices.columns)
print(returns.columns)
print(prices.isna().sum())
print(returns.isna().sum())
```

```python
prices_clean = prices.dropna()
returns_clean = prices_clean.pct_change().dropna()

print(prices_clean.shape)
print(returns_clean.shape)
print(prices_clean.head())
print(returns_clean.head())
```

```python
momentum = prices_clean.pct_change(63)

print(momentum.head())
print(momentum.tail())
print(momentum.shape)
print(momentum.isna().sum())
```

```python
volatility = returns_clean.rolling(63).std()

print(volatility.head())
print(volatility.tail())
print(volatility.shape)
print(volatility.isna().sum())
```

```python
cum = (1 + returns_clean).cumprod()
rolling_max = cum.cummax()
drawdown = (cum - rolling_max) / rolling_max

print(drawdown.head())
print(drawdown.tail())
print(drawdown.shape)
print(drawdown.min())
```

```python
common_index = momentum.index.intersection(volatility.index).intersection(drawdown.index)

momentum_aligned = momentum.loc[common_index]
volatility_aligned = volatility.loc[common_index]
drawdown_aligned = drawdown.loc[common_index]
```

```python
score = momentum_aligned - volatility_aligned + drawdown_aligned
print(score.head())
print(score.tail())
```

```python
sector_score = score.drop(columns=["SPY"])
sector_rank = sector_score.rank(axis=1, ascending=False)

print(sector_rank.tail())
```

```python
latest_rank = sector_rank.iloc[-1].sort_values()
print(latest_rank)
```
